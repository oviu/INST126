```python
ans = None

while True:
    choice = None
    while True:
        choice = input("Choose an operation: \n 1. Addition \n 2. Subtraction \n 3. Multiplication \n 4. Division \n 5. Exponentiation \n")
        if (choice == '1' or choice == '2' or choice == '3' or choice == '4' or choice == '5'):
            break
        print("Invalid input. Try again.")
        
    '''Each conditional if and elif branch denote a particular operation.
       For each operation if variable ans is None it means we are starting a new calculation, 
       if ans contains a value then we are on the same calculation and operating on that value'''
    
    if (choice == '1'):
        if (ans == None):
            print("Addition")
            n1 = float(input("Enter the first number: "))
            n2 = float(input("Enter the second number: "))
            ans = n1 + n2

        elif (ans != None):
            n3 = float(input("Enter number to add to " + str(ans) + " "))
            ans += n3

    elif (choice == '2'):
        if (ans == None):
            print("Subtraction")
            n1 = float(input("Enter the first number: "))
            n2 = float(input("Enter the number to subtract from " + str(n1) + ": " ))
            ans = n1 - n2
        
        elif (ans != None):
            n3 = float(input("Enter number to subtract from " + str(ans) + ": "))
            ans -= n3
            
    elif (choice == '3'):
        if (ans == None):
            print("Multiplication")
            n1 = float(input("Enter the first number: "))
            n2 = float(input("Enter the second number: "))
            ans = n1 * n2

        elif (ans != None):
            n3 = float(input("Enter number to multiply to " + str(ans) + ": "))
            ans *= n3
            
    elif (choice == '4'):
        if (ans == None):
            print("Division")
            n1 = float(input("Enter the first number: "))
            n2 = float(input("Enter the second number: "))
            ans = n1 / n2
            
        elif (ans != None):
            n3 = float(input("Divide " + str(ans) + " by: "))
            ans /= n3
            
    elif (choice == '5'):
        if (ans == None):
            print("Exponentiation")
            n1 = float(input("Enter the base number: "))
            n2 = float(input("Enter the exponent: "))
            ans = n1 ** n2
        
        elif (ans != None):
            n3 = float(input("Raise " + str(ans) + (" to the power: ")))
            ans **= n3
    else:
        print("Invalid Input")
        
    print("ans = " + str(ans))
    
    #logic for adding more calculations to current answer
    is_another = None
    while True:
        is_another = input("add another operation and operand to existing answer? (y/n): ")
        if (is_another == 'y' or is_another == 'n'):
            break
        print("Invalid input. Try again.")
        
    if (is_another == 'y'):
        continue
    elif (is_another == 'n'):
        pass
     
    #After a completed calculation. Logic for doing a new calculation or exiting program.
    is_continue = None
    while True:
        is_continue = input("Do another calculation? (y/n): ")
        if (is_continue == 'y' or is_continue == 'n'):
            break
        print("Invalid input. Try again.")

    if (is_continue == 'y'):
        ans = None
        continue
    elif (is_continue == 'n'):
        print("program exit.")
        break
```
